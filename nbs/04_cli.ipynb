```python
#| export
import sys
from fastcore.script import call_parse
from litesearch.data import mv_skill_md
```

```python
#| export
@call_parse
def install(
    dir: str = None  # target repo root (default: cwd)
):
    'Install SKILL.md to .agents/, .claude/, .Codex/ skill directories.'
    mv_skill_md(dry_run=False, dir=dir)

CMDS = {'install': install}

def main():
    if len(sys.argv) < 2 or sys.argv[1] not in CMDS:
        print('Usage: litesearch <command>')
        print('Commands: ' + ', '.join(CMDS))
        sys.exit(0)
        return
    cmd = sys.argv.pop(1)
    CMDS[cmd]()
```
